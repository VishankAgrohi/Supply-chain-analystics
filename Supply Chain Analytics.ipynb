````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("SCMS_Delivery_History_Dataset.csv")

# Convert date columns to datetime
date_cols = ['Delivered to Client Date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Clean numeric columns (remove commas, convert to float)
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'].astype(str).str.replace(",", ""), errors='coerce')
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'].astype(str).str.replace(",", ""), errors='coerce')
df['Line Item Value'] = pd.to_numeric(df['Line Item Value'], errors='coerce')
df['Line Item Quantity'] = pd.to_numeric(df['Line Item Quantity'], errors='coerce')

# Optional: Drop rows with missing key values for plotting
df.dropna(subset=["Line Item Value", "Delivered to Client Date"], inplace=True)

# Set style
sns.set(style="whitegrid")
````

#### Chart - 11- Correlation Heatmap

````python
# Correlation Heatmap visualization code
# Select only relevant numeric columns
numeric_cols = ['Line Item Quantity', 'Line Item Value', 'Pack Price', 'Unit Price',
                'Weight (Kilograms)', 'Freight Cost (USD)', 'Line Item Insurance (USD)']

# Convert Line Item Insurance column to numeric (if needed)
df['Line Item Insurance (USD)'] = pd.to_numeric(df['Line Item Insurance (USD)'].astype(str).str.replace(",", ""), errors='coerce')

# Compute correlation matrix
corr_matrix = df[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap of Key Numerical Features")
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer To understand how different numerical features relate to each other ‚Äî e.g., does increasing weight increase cost?

##### 2. What is/are the insight(s) found from the chart?

Answer Unit Price, Pack Price, and Line Item Value show strong positive correlation.

Freight Cost has moderate correlation with Weight and Line Item Value, implying heavier and more valuable shipments cost more to move.



#### Chart - 12 - Pair Plot

````python
# Pair Plot visualization code
# Filter the same numeric columns for pair plot
pairplot_df = df[numeric_cols].dropna()

# Pair plot
sns.pairplot(pairplot_df, diag_kind='kde', corner=True, plot_kws={'alpha': 0.5})
plt.suptitle("Pair Plot of Key Shipment Features", y=1.02)
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer To visually analyze distributions and pairwise relationships (e.g., clusters, outliers, linear trends) among critical cost/quantity features.

##### 2. What is/are the insight(s) found from the chart?

Answer Line Item Value increases with both Unit Price and Quantity.

Freight Cost appears more dispersed, with some outliers at low value but high cost (loss-makers).

Linear relationships visible between pricing metrics (Pack Price, Unit Price).
